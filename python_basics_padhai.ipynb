```
def print_square(stop_val):
  for i in range(stop_val):
    print(i, i**2)
```

```
def sum_of_squares(stop_val):
  sum_ = 0
  for i in range(stop_val):
    sum_ += i ** 2
  return sum_  
```

```
sum_5 = sum_of_squares(5)
```

```
def print_squares_with_startval(stop_val, start_val = 0): # if = 0 not specified then = 0 taken else taken whats fiven in the argument
  for i in range(start_val, stop_val):
    print(i, i**2)
```

```
print_squares_with_startval(5)
```

```
print_squares_with_startval(5,2)
```

```
class MobilePhone:
  '''blah blah'''
  def __init__(self, name, is_android = False, screen_size = 4.3):
    self.name = name
    self.is_android = is_android
    self.screen_size = screen_size
    self.rating = -1
  def has_rating(self):
    return self.rating >-1
    
```

```
def factorial(x):
  if x==0 :
    return 1
  return x* factorial(x-1)
```

```
factorial(5)
```
